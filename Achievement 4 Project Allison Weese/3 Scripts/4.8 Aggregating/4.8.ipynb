### 1.6 create spending flag

```python
# Calculate the mean price of products purchased by each user
user_avg_price = df.groupby('user_id')['prices'].mean().reset_index()

# Create a spending flag based on the average price
user_avg_price['spending_flag'] = user_avg_price['prices'].apply(lambda x: 'Low spender' if x < 10 else 'High spender')

# Merge the spending flag back into the original DataFrame
df = df.merge(user_avg_price[['user_id', 'spending_flag']], on='user_id', how='left')

# Display the DataFrame with the new spending_flag column
print(df.head())
```

```python
# Check
df.head()
```

### 1.7 Create order frequency flag

```python
# Calculate the median of 'days_since_prior_order' for each user
user_order_frequency = df.groupby('user_id')['days_since_prior_order'].median().reset_index()

# Create an order frequency flag based on the median value
def order_frequency_flag(median_days):
    if median_days > 20:
        return 'Non-frequent customer'
    elif 10 < median_days <= 20:
        return 'Regular customer'
    else:
        return 'Frequent customer'

user_order_frequency['order_frequency_flag'] = user_order_frequency['days_since_prior_order'].apply(order_frequency_flag)

# Merge the order frequency flag back into the original DataFrame
df = df.merge(user_order_frequency[['user_id', 'order_frequency_flag']], on='user_id', how='left')
```

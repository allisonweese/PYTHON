#### 1.4 Combine cust data with rest of instacart data

```python
# Ensure key columns are same data type
df_cust2['user_id'] = df_cust2['user_id'].astype(int)  # or str if needed
df_insta['user_id'] = df_insta['user_id'].astype(int)  # or str if needed
```

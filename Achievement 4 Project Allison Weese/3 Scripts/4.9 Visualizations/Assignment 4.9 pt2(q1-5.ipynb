```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Import ic df
df = pd.read_pickle(r'/Users/allisonweese/Documents/Achievement 4 Project Allison Weese/2 Data/Prepared Data/orders_products_combined.pkl')
```

```python
# Check head
df.head()
```

```python
# Assuming df is your DataFrame and 'order_hour_of_day' is the column you want to plot
hist = df['order_hour_of_day'].value_counts().sort_index().plot.bar()
# Adding labels to the x and y axes
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
# Adding a title (optional)
plt.title('Orders by Hour of the Day')
```

```python
# This histogram shows the busiest hours of the day on instacart
# The busiest hour is 10 and the least busy hour is 3
# Export hist
hist.figure.savefig(os.path.join('/Users/allisonweese/Documents/Achievement 4 Project Allison Weese/4 Analysis/Visualizations/historderhourofday.png'))
```

```python
# Create loyalty flag bar chart
bar = df['loyalty_flag'].value_counts().plot.bar()
# Adding labels to the x and y axes
plt.xlabel('Loyalty Flag')
plt.ylabel('Count')
# Adding a title (optional)
plt.title('Count of Loyalty Flags')
```

```python
# Regular/average customers purchase the most on ic
# Export bar chart
bar.figure.savefig(os.path.join('/Users/allisonweese/Documents/Achievement 4 Project/4 Analysis/Visualizations/loyaltyflagbar.png'))
```

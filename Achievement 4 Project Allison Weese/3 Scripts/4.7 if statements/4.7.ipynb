##### 1.3 Create and Use Price Label

```python
# Define function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply function
df['price_range'] = df.apply(price_label, axis=1)
```

### 2. Assignment

##### 2.1 Two Busiest and Two Slowest days

```python
# Calculate the order counts for each day
order_counts = ords_prods_merge['order_dow'].value_counts()
```

```python
# Identify the top two busiest days
busiest_days = order_counts.nlargest(2).index.tolist()
```

```python
# Identify the two slowest days
slowest_days = order_counts.nsmallest(2).index.tolist()
```

```python
# Create a new column based on these identified days
def label_day(day):
    if day in busiest_days:
        return 'Busiest days'
    elif day in slowest_days:
        return 'Slowest days'
    else:
        return 'Regular day'

ords_prods_merge['day_label'] = ords_prods_merge['order_dow'].apply(label_day)
```

```python
# Print the resulting value counts for verification
print("Value counts for the new 'day_label' column:")
print(ords_prods_merge['day_label'].value_counts())
```

```python
# Check slowest days
slowest_days
```

```python
# Check busiest days
busiest_days
```

```python
# The slowest days are Wednesdays and Tuesdays
# The Busiest days are Sundays and Mondays
# Thursday, Friday, and Saturday are regular days
```

##### 2.2 Busiest times of day

```python
# Calculate the order counts for each hour of the day
hourly_order_counts = ords_prods_merge.groupby('order_hour_of_day').size()
```

```python
# Print order counts
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Grouping hours
result = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result.append("Most Orders")
    elif value in [3, 4, 2, 5, 1, 0, 6, 23]:
        result.append("Fewest Orders")
    else:
        result.append("Average Orders")
```

```python
# Create Result
ords_prods_merge['busiest_hour'] = result
```

```python
# Print Result
ords_prods_merge['busiest_hour'].value_counts(dropna = False)
```
